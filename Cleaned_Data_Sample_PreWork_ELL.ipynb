```python
import pandas as pd
```

```python
bike_file = "./data/JC-201901-1-citibike-tripdata.csv"
bike_df = pd.read_csv(bike_file)
```

```python
bike_df.head()
```

```python
bike_df.shape[0]
```

```python
reduced_bike_df = bike_df.sample(n=10000).reset_index(drop=True)
reduced_bike_df
```

```python
def genderID (row):
    if row['gender'] == 0:
        return 'Unknown'
    if row['gender'] == 1:
        return 'Male'
    if row['gender'] == 2:
        return 'Female'

reduced_bike_df['Gender'] = reduced_bike_df.apply(lambda row: genderID(row), axis=1)

reduced_bike_df.head()
        
```

```python
new = reduced_bike_df['starttime'].str.split(" ", n = 1, expand = True)
reduced_bike_df["Start Date"]= new[0]
reduced_bike_df["Start Time"]= new[1]
```

```python
new2 = reduced_bike_df['stoptime'].str.split(" ", n = 1, expand = True)
reduced_bike_df["Stop Date"]= new2[0]
reduced_bike_df["Stop Time"]= new2[1]
```

```python
reduced_bike_df.drop(columns =['starttime','stoptime'], inplace = True)
reduced_bike_df["Stop Time"] = reduced_bike_df["Stop Time"].map(lambda x: str(x)[:-8])
reduced_bike_df["Start Time"] = reduced_bike_df["Start Time"].map(lambda x: str(x)[:-8])
```

```python
new3 = reduced_bike_df['Start Time'].str.split(":", n = 1, expand = True)
reduced_bike_df["Start Hour"]= new3[0]
reduced_bike_df["Start Minute"]= new3[1]

new4 = reduced_bike_df['Stop Time'].str.split(":", n = 1, expand = True)
reduced_bike_df["Stop Hour"]= new4[0]
reduced_bike_df["Stop Minute"]= new4[1]

reduced_bike_df.drop(columns =['Start Time','Stop Time'], inplace = True)
```

```python
new5 = reduced_bike_df['Start Date'].str.split("-", n = 2, expand = True)
reduced_bike_df["Start Year"]= new5[0]
reduced_bike_df["Start Month"]= new5[1]
reduced_bike_df["Start Day"]= new5[2]
```

```python
new6 = reduced_bike_df['Stop Date'].str.split("-", n = 2, expand = True)
reduced_bike_df["Stop Year"]= new6[0]
reduced_bike_df["Stop Month"]= new6[1]
reduced_bike_df["Stop Day"]= new6[2]
```

```python
reduced_bike_df.head()
```

```python
reduced_bike_df.columns
```

```python
cleaned_bike_df = reduced_bike_df[['Start Date','Stop Date', 'Start Hour', 'Start Minute', 'Stop Hour', 'Stop Minute','Start Year', 'Start Month', 'Start Day', 'Stop Year', 'Stop Month','Stop Day','tripduration', 'start station id', 'start station name','start station latitude', 'start station longitude', 'end station id','end station name', 'end station latitude', 'end station longitude','bikeid', 'usertype', 'birth year','Gender']]
```

```python
cleaned_bike_df.head()
```

```python
cleaned_bike_df.to_csv("cleaned_10k_bike_data.csv")
```
